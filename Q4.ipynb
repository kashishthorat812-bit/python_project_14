```python
import numpy as np
# create a 1D array between 1 to 24 using arange()
a1=np.arange(1,25)
print("1D Array :")
print(a1)

# reshape the 1D array into (4,6) matrix
a2=a1.reshape(4,6)
print("\nReshape the 1D array into (4,6) matrix :")
print(a2)
print("Shape = ",np.shape(a2))

# reshape the 1D array into (3,8) matrix
b=a1.reshape(4,6)
print("\nReshape the 1D array into (3,8) matrix :")
print(b)
print("Shape = ",np.shape(b))

# reshape the 1D array into (2,3,4) 3D Array
c=a1.reshape(2,3,4)
print("\nReshape the 1D array into (2,3,4) 3D Array :")
print(c)
print("Shape = ",np.shape(c))
```
