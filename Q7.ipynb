```python
import numpy as np
a=np.array([10,20,30,40])
b=np.array([1,2,3,4])
print("Array a = ",a)
print("Array b = ",b)

# Addition of two array
c=a+b
print("\nAddition :")
print(c)

# Substraction of two array
c1=a-b
print("\nSubstraction :")
print(c1)

# Multiplication of two array
c2=a*b
print("\nMultiplication :")
print(c2)

# Division of two array
c3=a/b
print("\nDivision :")
print(c3)

# Power of  array
c4=a**b
print("\nPower :")
print(c4)

# Modul0 of two array
c5=a%b
print("\nModulo :")
print(c5)

```
