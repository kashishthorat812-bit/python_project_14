```python
import numpy as np
# create a 1D array with random integers between 1 to 50
a=np.random.randint(1,51,20)
print("1D Array :")
print(a)

# print the minimum value and its index using (min) and (argmin)
print("\nMinimum Value = ",np.min(a))
print("Index of minimum value = ",np.argmin(a))

# print the Maximum value and its index using (max) and (argmax)
print("\nMaximum Value = ",np.max(a))
print("Index of Maximum value = ",np.argmax(a))

# print the sum of element using (sum)
print("\nSum of Element = ",np.sum(a))

# print the mean using (mean))
print("Mean of 1D Array = ",np.mean(a))

# print the Standard Deviation using (std)
print("Standard Deviation = ",np.std(a))
```
