```python
import numpy as np

marks = np.random.randint(30, 101, (10, 5))

print("Student Marks Array:\n", marks)

total_marks = np.sum(marks, axis=1)
average_marks = np.mean(marks, axis=1)

print("\nTotal Marks of each student:", total_marks)
print("Average Marks of each student:", average_marks)

highest_student = np.argmax(total_marks)
lowest_student = np.argmin(total_marks)

print("\nStudent with highest total marks: Student", highest_student + 1)
print("Highest total marks:", total_marks[highest_student])

print("Student with lowest total marks: Student", lowest_student + 1)
print("Lowest total marks:", total_marks[lowest_student])

print("\nOverall class mean:", np.mean(marks))
print("Overall class standard deviation:", np.std(marks))

top3_indices = np.argsort(total_marks)[-3:][::-1]

print("\nTop 3 students indices:", top3_indices + 1)
print("Marks of top 3 students:\n", marks[top3_indices])
```
