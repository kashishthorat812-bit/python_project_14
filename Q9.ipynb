```python
import numpy as np

arr = np.random.randn(6, 6)

print("Original Matrix :")
print(arr)

print("\nShape :")
print(arr.shape)

print("\nSize :")
print(arr.size)

print("\nData type :")
print(arr.dtype)

print("Index of maximum value :")
print(np.unravel_index(np.argmax(arr), arr.shape))
print("Index of minimum value :")
print(np.unravel_index(np.argmin(arr), arr.shape))

print("Top-left 3x3 submatrix:\n", arr[:3, :3])

arr[arr < 0] = np.abs(arr[arr < 0])

print("Modified Matrix:\n", arr)
print("Mean of modified matrix:", np.mean(arr))
```
