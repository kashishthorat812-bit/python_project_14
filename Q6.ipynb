```python
import numpy as np
# create a 5*5 matrix of random number between
a=np.random.randint(1,101,(5,5))
print("Matrix :")
print(a)

# print the diagonal element using (diag)
print("\nDiagonal Element :")
print(np.diag(a))

# print the element greater than 50
print("\nElement greater tha 50 :")
print(a[a>50])

# print the modify matrix
a[a<30]=0
print("\nModify Matrix :")
print(a)
```
