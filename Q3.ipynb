```python
import numpy as np
# create a 4*5 matrix of random integers between 20 to 80
a=np.random.randint(20,81,(4,5))
print("Matrix :")
print(a)

# print the minimum and maximum value
print("\nMinimum Value = ",np.min(a))
print("Maximum Value = ",np.max(a))

# print the sum of all element using (sum)
print("\nSum of All Element = ",np.sum(a))

# print the mean using (mean)
print("Mean of matrix = ",np.mean(a))

# print standard deviation of matrix using (std)
print("Standard Deviation of matrix = ",np.std(a))

# print the sum of rows and column
print("\nSum of Rows = ",np.sum(a,axis=1))
print("Sum of Column = ",np.sum(a,axis=0))
```
